# Description Embeddings and t-SNE Dimensionality Reduction
Generate embeddings for job descriptions and reduce to 3D for visualization


```python
import numpy as np
from openai import OpenAI
from sklearn.manifold import TSNE
from dotenv import load_dotenv
import os
from tqdm import tqdm

# Load environment variables
load_dotenv()

# Initialize OpenAI client
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise ValueError("Please set OPENAI_API_KEY in .env file")
client = OpenAI(api_key=api_key)

print("Libraries imported successfully!")
```

```python
# Get embeddings for descriptions only
# Extract descriptions (handle NaN values)
descriptions = sampled_engineers_df['description'].fillna('').astype(str).tolist()
print(f"Total descriptions to embed: {len(descriptions)}")
print(f"Number of empty descriptions: {sum(1 for d in descriptions if not d.strip())}")

# Batch size for embedding API calls
batch_size = 32
embeddings = []

print(f"\nGenerating embeddings in batches of {batch_size}...")
for i in tqdm(range(0, len(descriptions), batch_size), desc="Embedding batches"):
    batch = descriptions[i:i + batch_size]
    try:
        response = client.embeddings.create(
            model="text-embedding-3-large",
            input=batch
        )
        batch_vectors = [r.embedding for r in response.data]
        embeddings.extend(batch_vectors)
    except Exception as e:
        print(f"Error processing batch {i//batch_size + 1}: {e}")
        # Add zero vectors as placeholders for failed batches
        embeddings.extend([[0.0] * 3072 for _ in batch])

embeddings_array = np.array(embeddings).astype("float32")
print(f"\nEmbeddings shape: {embeddings_array.shape}")
print(f"Expected dimension for text-embedding-3-large: 3072")
```

```python
# Sanity checks
assert len(sampled_engineers_df) == len(descriptions), f"Row/description mismatch: {len(sampled_engineers_df)} vs {len(descriptions)}"
assert embeddings_array.ndim == 2, f"Embeddings should be 2D, got {embeddings_array.ndim}D"
assert embeddings_array.shape[0] == len(sampled_engineers_df), f"Row/embedding mismatch: {len(sampled_engineers_df)} vs {embeddings_array.shape[0]}"

embedding_dim = embeddings_array.shape[1]
print(f"Attaching embeddings of dimension {embedding_dim} to DataFrame...")

# Attach embeddings as list-of-floats per row (more portable across formats)
sampled_engineers_df = sampled_engineers_df.copy()
sampled_engineers_df["description_embedding"] = [vec.tolist() for vec in embeddings_array]

# Save a copy with timestamped filename
ts = pd.Timestamp.now().strftime("%Y%m%d_%H%M%S")
base = f"sampled_engineers_with_embeddings_{embedding_dim}d_{ts}"
parquet_path = f"{base}.parquet"
csv_path = f"{base}.csv"

# Prefer Parquet (compact, preserves list type); also write CSV as a simple fallback
try:
    sampled_engineers_df.to_parquet(parquet_path, index=False)
    print(f"Saved Parquet: {parquet_path}")
except Exception as e:
    print(f"Parquet save failed ({e}). You may need 'pyarrow' or 'fastparquet' installed. Skipping Parquet.")

sampled_engineers_df.to_csv(csv_path, index=False)
print(f"Saved CSV: {csv_path}")

print("\nSample of the new column:")
print(sampled_engineers_df["description_embedding"].head(2).apply(lambda v: f"len={len(v)} first3={v[:3]}"))
```

```python
# HDBSCAN Clustering on Raw Embeddings for Small, Tight Groups
# This approach uses HDBSCAN directly on embeddings (not t-SNE) with strict parameters
# to find small groups of highly similar descriptions

import numpy as np
import hdbscan
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import connected_components
import pandas as pd

# Normalize embeddings for cosine metric
embeddings_normalized = embeddings_array / np.linalg.norm(embeddings_array, axis=1, keepdims=True)
```

```python
# Clean and filter normalized embeddings before HDBSCAN
import numpy as np
import pandas as pd

# Preconditions: embeddings_array (raw), embeddings_normalized (possibly with NaNs from 0/0)
raw = embeddings_array
normed = embeddings_normalized

# Diagnose issues
raw_norms = np.linalg.norm(raw, axis=1)
zero_raw = (raw_norms == 0)

nan_raw = np.isnan(raw).any(axis=1)
inf_raw = np.isinf(raw).any(axis=1)

nan_normed = np.isnan(normed).any(axis=1)
inf_normed = np.isinf(normed).any(axis=1)

bad_mask = zero_raw | nan_raw | inf_raw | nan_normed | inf_normed
keep_mask = ~bad_mask

kept_indices = np.where(keep_mask)[0]
dropped_indices = np.where(bad_mask)[0]

print(f"Total rows: {normed.shape[0]}")
print(f"Dropping: {bad_mask.sum()} rows")
print(
    " - zero-norm raw:", int(zero_raw.sum()),
    "| NaN raw:", int(nan_raw.sum()),
    "| Inf raw:", int(inf_raw.sum()),
    "| NaN normed:", int(nan_normed.sum()),
    "| Inf normed:", int(inf_normed.sum())
)

# Build clean arrays and optional aligned DataFrame slice
embeddings_normalized_clean = normed[keep_mask].astype(np.float32, copy=False)

# Re-normalize safely (should be unnecessary for most, but ensures unit length/stability)
norms_clean = np.linalg.norm(embeddings_normalized_clean, axis=1, keepdims=True)
embeddings_normalized_clean = embeddings_normalized_clean / np.maximum(norms_clean, 1e-12)

# Optional: subset your DataFrame to keep alignment and remember original indices
try:
    df_hdbscan = sampled_engineers_df.loc[kept_indices].copy()
    df_hdbscan["orig_index"] = kept_indices
    print(f"Kept DataFrame rows: {len(df_hdbscan)}")
except NameError:
    df_hdbscan = None
    print("Note: 'sampled_engineers_df' not found; skipping DataFrame subset.")

# Final safety checks
assert np.isfinite(embeddings_normalized_clean).all(), "Non-finite values remain after filtering."
print("embeddings_normalized_clean ready for HDBSCAN. Use kept_indices to map back to originals.")
```

```python

# Step 1: HDBSCAN with strict parameters for small, tight clusters
print("Running HDBSCAN on embeddings...")
clusterer = hdbscan.HDBSCAN(
    min_cluster_size=3,           # Small minimum cluster size
    min_samples=2,                # Strict min_samples (must be < min_cluster_size)
    metric='euclidean',              # Cosine metric for semantic similarity
    cluster_selection_method='leaf',  # Leaf selection for granular clusters
    prediction_data=True          # Enable probability calculation
)

labels = clusterer.fit_predict(embeddings_normalized_clean)
probabilities = clusterer.probabilities_

print(f"Initial clusters found: {len(set(labels)) - (1 if -1 in labels else 0)}")
print(f"Noise points: {np.sum(labels == -1)}")
```

```python
# Step 2: Probability filtering - keep only high-confidence members
PROB_THRESHOLD = 0.5  # Only keep members with >50% probability
filtered_labels = labels.copy()
filtered_labels[probabilities < PROB_THRESHOLD] = -1

print(f"\nAfter probability filtering (threshold={PROB_THRESHOLD}):")
print(f"Clusters remaining: {len(set(filtered_labels)) - (1 if -1 in filtered_labels else 0)}")
```

```python
# Step 3: Post-split large clusters (>15 members) using high-similarity graph
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import connected_components

MAX_CLUSTER_SIZE = 15
SPLIT_SIMILARITY_THRESHOLD = 0.92  # High similarity threshold for splitting

# Preconditions
assert embeddings_normalized_clean.shape[0] == len(filtered_labels), \
    f"Mismatch: {embeddings_normalized_clean.shape[0]} vs {len(filtered_labels)}"

final_labels = filtered_labels.copy()

# Initialize next_cluster_id safely (handle the all -1 case)
if np.any(final_labels >= 0):
    next_cluster_id = int(np.max(final_labels[final_labels >= 0])) + 1
else:
    next_cluster_id = 0

for cluster_id in np.unique(filtered_labels):
    if cluster_id == -1:
        continue

    cluster_mask = (final_labels == cluster_id)  # mask in the CLEAN space
    cluster_size = int(cluster_mask.sum())
    if cluster_size <= MAX_CLUSTER_SIZE:
        continue

    # Indices in the CLEAN space
    cluster_indices_clean = np.where(cluster_mask)[0]
    cluster_embeddings = embeddings_normalized_clean[cluster_indices_clean]

    # Compute pairwise cosine similarity WITHIN this cluster
    sim_matrix = cosine_similarity(cluster_embeddings)

    # Build adjacency at high similarity threshold
    adjacency = (sim_matrix >= SPLIT_SIMILARITY_THRESHOLD).astype(np.int8)
    np.fill_diagonal(adjacency, 0)  # no self loops

    # Connected components => subclusters
    n_components, component_labels = connected_components(
        csgraph=csr_matrix(adjacency), directed=False
    )

    if n_components > 1:
        print(f"Splitting cluster {cluster_id} (size {cluster_size}) into {n_components} sub-clusters")
        for comp_id in range(n_components):
            comp_mask = (component_labels == comp_id)
            global_clean_idx = cluster_indices_clean[comp_mask]
            if comp_id == 0:
                final_labels[global_clean_idx] = cluster_id
            else:
                final_labels[global_clean_idx] = next_cluster_id
                next_cluster_id += 1

print(f"\nAfter splitting large clusters:")
print(f"Total clusters: {len(set(final_labels)) - (1 if -1 in final_labels else 0)}")

# Optional: map back to original indices (if you need to align with the full DataFrame)
try:
    final_labels_full = np.full(embeddings_array.shape[0], -1, dtype=int)
    final_labels_full[kept_indices] = final_labels
    print("Built final_labels_full aligned to original rows via kept_indices.")
except NameError:
    pass
```

```python
# Step 4: Gate clusters by intra-extra similarity margin (CLEAN space)
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

MARGIN_THRESHOLD = 0.15  # Intra-cluster similarity must exceed extra-cluster by this margin

# Ensure alignment and array types
final_labels = np.asarray(final_labels)
assert embeddings_normalized_clean.shape[0] == len(final_labels), \
    f"Mismatch: {embeddings_normalized_clean.shape[0]} vs {len(final_labels)}"

gated_labels = final_labels.copy()
clusters_to_check = [c for c in np.unique(final_labels) if c != -1]

print(f"\nApplying intra-extra similarity margin gate (threshold={MARGIN_THRESHOLD})...")
removed_clusters = []

for cluster_id in clusters_to_check:
    # Masks and indices in CLEAN space
    cluster_mask = (final_labels == cluster_id)
    cluster_indices_clean = np.where(cluster_mask)[0]

    # Remove singleton clusters
    if cluster_indices_clean.size < 2:
        gated_labels[cluster_indices_clean] = -1
        removed_clusters.append(cluster_id)
        continue

    non_cluster_mask = ~cluster_mask
    non_cluster_indices_clean = np.where(non_cluster_mask)[0]

    cluster_embeddings = embeddings_normalized_clean[cluster_indices_clean]
    non_cluster_embeddings = embeddings_normalized_clean[non_cluster_indices_clean]

    # If no outside points, skip gating for this cluster
    if non_cluster_embeddings.size == 0:
        continue

    # Intra-cluster similarity (exclude diagonal)
    intra_sim_matrix = cosine_similarity(cluster_embeddings)
    iu = np.triu_indices(cluster_embeddings.shape[0], k=1)
    if iu[0].size == 0:
        # Fallback if only two points
        intra_similarity = float(intra_sim_matrix[~np.eye(intra_sim_matrix.shape[0], dtype=bool)].mean())
    else:
        intra_similarity = float(intra_sim_matrix[iu].mean())

    # Extra-cluster similarity (cluster vs all non-cluster points)
    extra_sim_matrix = cosine_similarity(cluster_embeddings, non_cluster_embeddings)
    extra_similarity = float(extra_sim_matrix.mean())

    # Margin check
    margin = intra_similarity - extra_similarity
    if margin < MARGIN_THRESHOLD:
        gated_labels[cluster_indices_clean] = -1
        removed_clusters.append(cluster_id)

print(f"Removed {len(removed_clusters)} clusters due to insufficient separation")
print(f"Final clusters: {len(set(gated_labels)) - (1 if -1 in gated_labels else 0)}")

# Summary statistics
final_cluster_ids = [c for c in np.unique(gated_labels) if c != -1]
cluster_sizes = [int((gated_labels == c).sum()) for c in final_cluster_ids]

if cluster_sizes:
    print("\nCluster size statistics:")
    print(f"  Mean: {np.mean(cluster_sizes):.1f}")
    print(f"  Median: {np.median(cluster_sizes):.1f}")
    print(f"  Min: {np.min(cluster_sizes)}")
    print(f"  Max: {np.max(cluster_sizes)}")
    print(f"  Total points in clusters: {int((gated_labels != -1).sum())}")
    print(f"  Noise/ungrouped points: {int((gated_labels == -1).sum())}")

    print("\nCluster size distribution:")
    for size in sorted(set(cluster_sizes)):
        count = cluster_sizes.count(size)
        print(f"  Size {size}: {count} cluster(s)")
else:
    print("\nNo clusters found meeting all criteria.")

# Store results for further analysis (probabilities may come from HDBSCAN; handle if missing)
try:
    probs_out = probabilities
except NameError:
    probs_out = None

cluster_results = {
    "labels": gated_labels,
    "probabilities": probs_out,
    "n_clusters": len(final_cluster_ids),
    "cluster_sizes": cluster_sizes,
}

# Optional: map back to original rows via kept_indices (if you need full-array alignment)
try:
    gated_labels_full = np.full(embeddings_array.shape[0], -1, dtype=int)
    gated_labels_full[kept_indices] = gated_labels
    print("Built gated_labels_full aligned to original rows via kept_indices.")
except NameError:
    pass
```

```python
# Attach cluster info back to sampled_engineers_df and save
import numpy as np
import pandas as pd

# 1) Pick the best available labels/probabilities
try:
    labels_in = gated_labels  # preferred: after gating
except NameError:
    try:
        labels_in = final_labels  # fallback: before gating
    except NameError:
        labels_in = labels  # last resort: raw HDBSCAN labels

try:
    probs_in = probs_out  # from gating cell (optional)
except NameError:
    try:
        probs_in = probabilities  # from HDBSCAN fit (optional)
    except NameError:
        probs_in = None

labels_in = np.asarray(labels_in)
n_df = len(sampled_engineers_df)

# 2) Build full-length arrays aligned to the original DataFrame rows
labels_full = np.full(n_df, -1, dtype=int)
probs_full = np.full(n_df, np.nan, dtype=np.float32)

if labels_in.shape[0] == n_df:
    # Already aligned to the full dataset
    labels_full = labels_in.astype(int, copy=False)
    if probs_in is not None and len(probs_in) == n_df:
        probs_full = np.asarray(probs_in, dtype=np.float32)
    align_msg = "Aligned directly to full DataFrame."
else:
    # Expect a cleaned-subset alignment via kept_indices
    try:
        idx = np.asarray(kept_indices)
    except NameError as e:
        raise ValueError(
            "Labels length does not match DataFrame and 'kept_indices' is not available "
            "to map cleaned subset back to original rows."
        ) from e

    if labels_in.shape[0] != idx.shape[0]:
        raise ValueError(
            f"Label count ({labels_in.shape[0]}) does not match kept_indices ({idx.shape[0]})."
        )

    labels_full[idx] = labels_in.astype(int, copy=False)
    if probs_in is not None:
        if len(probs_in) != idx.shape[0]:
            print(f"Warning: probabilities length ({len(probs_in)}) does not match kept_indices ({idx.shape[0]}). Skipping probabilities.")
        else:
            probs_full[idx] = np.asarray(probs_in, dtype=np.float32)
    align_msg = "Aligned via kept_indices to original DataFrame."

print(align_msg)
print(f"Final label counts (incl. noise=-1): {dict(zip(*np.unique(labels_full, return_counts=True)))}")

# 3) Attach to DataFrame
df_out = sampled_engineers_df.copy()
df_out["cluster_label"] = labels_full
df_out["cluster_probability"] = probs_full
df_out["clustered"] = (df_out["cluster_label"] != -1)

# 4) Save with timestamp
ts = pd.Timestamp.now().strftime("%Y%m%d_%H%M%S")
base = f"sampled_engineers_with_clusters_{ts}"
parquet_path = f"{base}.parquet"
csv_path = f"{base}.csv"

try:
    df_out.to_parquet(parquet_path, index=False)
    print(f"Saved Parquet: {parquet_path}")
except Exception as e:
    print(f"Parquet save failed ({e}). You may need 'pyarrow' or 'fastparquet'. Skipping Parquet.")

df_out.to_csv(csv_path, index=False)
print(f"Saved CSV: {csv_path}")

# 5) Quick peek
print("Preview:", df_out[["cluster_label", "cluster_probability", "clustered"]].head(5))
```

```python
# Load cluster_df, (re)build or reuse reps table, then resume-safe CV generation with timeout + checkpoints
import os, sys, json, time, glob, concurrent.futures
import pandas as pd
from tqdm import tqdm

# --- CONFIG ---
PARQUET_PATH = "sampled_engineers_with_clusters_20251105_175242.parquet"
TIMEOUT_SEC = 120           # per-call timeout
OUTER_MAX_ATTEMPTS = 2      # 1 try + 1 retry
CHECKPOINT_EVERY = 5        # save progress every N reps
CHECKPOINT_PREFIX = "cluster_reps_checkpoint"  # checkpoint filename prefix
REQUIRE_MIN_DESC_LEN = 200  # representatives must have description length >= 200

# --- IMPORTS ---
if "src" not in sys.path:
    sys.path.append("src")
from cv_generator import generate_cv_for_job  # returns (cv_text, info)

# --- LOAD CLUSTER DF ---
print(f"Loading cluster_df from: {PARQUET_PATH}")
cluster_df = pd.read_parquet(PARQUET_PATH)
assert "description" in cluster_df.columns, "cluster_df must contain a 'description' column."

# Detect cluster label column
possible_label_cols = ["cluster_label", "gated_labels", "final_labels", "labels", "cluster", "cluster_id", "label"]
label_col = next((c for c in possible_label_cols if c in cluster_df.columns), None)
if label_col is None:
    raise KeyError(f"Could not find a cluster label column. Tried: {possible_label_cols}")

# Columns typically used by the generator; include those present
gen_cols = ["title_raw", "company", "description", "job_category", "role_k50"]
present_gen_cols = [c for c in gen_cols if c in cluster_df.columns]

# --- BUILD FRESH REPRESENTATIVES TABLE (one row per cluster, exclude -1, require min desc len) ---
def _select_rep_index(group: pd.DataFrame) -> int:
    # Filter to rows with description length >= threshold
    desc_len = group["description"].fillna("").astype(str).str.len()
    eligible = group[desc_len >= REQUIRE_MIN_DESC_LEN]
    if eligible.empty:
        return None
    if "cluster_probability" in eligible.columns and eligible["cluster_probability"].notna().any():
        return int(eligible["cluster_probability"].astype(float).idxmax())
    # Fallback: longest description
    return int(eligible["description"].fillna("").astype(str).str.len().idxmax())

rows = []
skipped_clusters = []
for cid, group in cluster_df[cluster_df[label_col] != -1].groupby(label_col):
    rep_idx = _select_rep_index(group)
    if rep_idx is None:
        skipped_clusters.append(cid)
        continue
    row = {"cluster_id": int(cid) if pd.notna(cid) else cid, "row_index": rep_idx}
    for c in present_gen_cols:
        row[c] = cluster_df.at[rep_idx, c]
    rows.append(row)

fresh_reps_df = pd.DataFrame(rows).sort_values("cluster_id").reset_index(drop=True)
print(f"Fresh representatives table: {len(fresh_reps_df)} clusters (skipped {len(skipped_clusters)} with desc < {REQUIRE_MIN_DESC_LEN}).")

# --- TRY TO LOAD LATEST CHECKPOINT; ELSE USE FRESH ---
def _load_latest_checkpoint(prefix: str):
    pq = sorted(glob.glob(f"{prefix}_*.parquet"), key=os.path.getmtime)
    cs = sorted(glob.glob(f"{prefix}_*.csv"), key=os.path.getmtime)
    latest = pq[-1] if pq else (cs[-1] if cs else None)
    if not latest:
        return None, None
    try:
        df = pd.read_parquet(latest) if latest.endswith(".parquet") else pd.read_csv(latest)
        return df, latest
    except Exception as e:
        print(f"Failed to load checkpoint {latest}: {e}")
        return None, None

reps_df_ckpt, ckpt_path = _load_latest_checkpoint(CHECKPOINT_PREFIX)
if reps_df_ckpt is not None:
    # Use checkpoint as authoritative reps_df
    reps_df = reps_df_ckpt.copy()
    print(f"Loaded reps checkpoint: {ckpt_path} (rows={len(reps_df)})")
else:
    reps_df = fresh_reps_df.copy()
    print("No reps checkpoint found; using fresh representatives.")

# Ensure mandatory columns exist on reps_df
for col in [
    "cv_standard", "cv_error_standard", "cv_extracted_requirements_standard",
    "cv_desc_only", "cv_error_desc", "cv_extracted_requirements_desc"
]:
    if col not in reps_df.columns:
        reps_df[col] = pd.NA

# --- REPORT CURRENT COMPLETION ---
has_std = reps_df["cv_standard"].notna()
has_desc = reps_df["cv_desc_only"].notna()
done_both = (has_std & has_desc).sum()
print(f"Current reps status: std={has_std.sum()}, desc_only={has_desc.sum()}, both={done_both}, total={len(reps_df)}")

# --- PREPARE GENERATION ---
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    print("Note: OPENAI_API_KEY not found; .env may be used if available.")
to_process = reps_df.index[reps_df["cv_standard"].isna() | reps_df["cv_desc_only"].isna()].tolist()
print(f"Generating CVs for {len(to_process)} reps (timeout {TIMEOUT_SEC}s, 1 retry)...")

def _call_with_timeout(job_row: dict, desc_only: bool, timeout: int):
    def _invoke():
        return generate_cv_for_job(job_row, api_key=api_key, description_only=desc_only)
    with concurrent.futures.ThreadPoolExecutor(max_workers=1) as ex:
        fut = ex.submit(_invoke)
        return fut.result(timeout=timeout)

def _gen_for_row(idx: int, desc_only: bool):
    cv_col = "cv_standard" if not desc_only else "cv_desc_only"
    err_col = "cv_error_standard" if not desc_only else "cv_error_desc"
    req_col = "cv_extracted_requirements_standard" if not desc_only else "cv_extracted_requirements_desc"
    if not pd.isna(reps_df.at[idx, cv_col]):
        return  # already done

    # Build minimal job_row payload
    job_row = {c: reps_df.at[idx, c] for c in gen_cols if c in reps_df.columns}

    last_err = None
    for attempt in range(OUTER_MAX_ATTEMPTS):
        try:
            cv_text, info = _call_with_timeout(job_row, desc_only=desc_only, timeout=TIMEOUT_SEC)
            reps_df.at[idx, cv_col] = cv_text
            reps_df.at[idx, req_col] = json.dumps(info.get("extracted_requirements", {}), ensure_ascii=False)
            reps_df.at[idx, err_col] = pd.NA
            return
        except Exception as e:
            last_err = e
            if attempt < OUTER_MAX_ATTEMPTS - 1:
                time.sleep(2)  # small backoff
    reps_df.at[idx, err_col] = f"timeout_or_error: {last_err}"

# --- RUN GENERATION WITH PERIODIC CHECKPOINTS ---
processed = 0
for i in tqdm(to_process, desc="Reps"):
    _gen_for_row(i, desc_only=False)
    _gen_for_row(i, desc_only=True)

    processed += 1
    if processed % CHECKPOINT_EVERY == 0:
        ts = pd.Timestamp.now().strftime("%Y%m%d_%H%M%S")
        fn = f"{CHECKPOINT_PREFIX}_{ts}.parquet"
        try:
            reps_df.to_parquet(fn, index=False)
            print(f"Checkpoint saved: {fn}")
        except Exception as e:
            print(f"Checkpoint Parquet failed ({e}); saving CSV fallback...")
            reps_df.to_csv(f"{CHECKPOINT_PREFIX}_{ts}.csv", index=False)
            print(f"Checkpoint saved (CSV): {CHECKPOINT_PREFIX}_{ts}.csv")

# --- FINAL SUMMARY + SAVE FINAL CHECKPOINT ---
remaining = int((reps_df["cv_standard"].isna() | reps_df["cv_desc_only"].isna()).sum())
done_std = int(reps_df["cv_standard"].notna().sum())
done_desc = int(reps_df["cv_desc_only"].notna().sum())
print(f"Complete. std={done_std}, desc_only={done_desc}, remaining pairs={remaining}")

ts = pd.Timestamp.now().strftime("%Y%m%d_%H%M%S")
final_fn = f"{CHECKPOINT_PREFIX}_final_{ts}.parquet"
try:
    reps_df.to_parquet(final_fn, index=False)
    print(f"Final checkpoint saved: {final_fn}")
except Exception as e:
    print(f"Final Parquet failed ({e}); saving CSV fallback.")
    reps_df.to_csv(f"{CHECKPOINT_PREFIX}_final_{ts}.csv", index=False)
    print(f"Final checkpoint saved (CSV): {CHECKPOINT_PREFIX}_final_{ts}.csv")
```

```python
import numpy as np
import pandas as pd

def inspect_random_cluster(
    cluster_df: pd.DataFrame,
    cluster_id: int | None = None,
    label_col: str | None = None,
    sample_n: int = 2,
    seed: int | None = None,
):
    """
    Inspect one cluster to sanity-check coherence.
    - Picks a random cluster (excluding -1) unless cluster_id is provided.
    - Prints value counts for key fields and two random descriptions.
    - Returns (chosen_cluster_id, cluster_slice) for further exploration.
    """
    # Pick label column if not provided
    if label_col is None:
        candidates = ["cluster_label", "gated_labels", "final_labels", "labels", "cluster", "cluster_id", "label"]
        label_col = next((c for c in candidates if c in cluster_df.columns), None)
        if label_col is None:
            raise KeyError(f"Could not find a cluster label column. Tried: {candidates}")

    # Collect valid cluster ids (exclude -1)
    valid_ids = [cid for cid in cluster_df[label_col].dropna().unique().tolist() if cid != -1]
    if not valid_ids:
        raise ValueError("No non-noise clusters found (all are -1).")

    # Choose cluster id
    if cluster_id is None:
        rng = np.random.default_rng(seed)
        cluster_id = rng.choice(valid_ids)
    else:
        if cluster_id not in valid_ids:
            raise ValueError(f"cluster_id {cluster_id} not found among non-noise clusters: {sorted(valid_ids)[:10]}...")

    # Slice the cluster
    mask = (cluster_df[label_col] == cluster_id)
    cdf = cluster_df.loc[mask]
    size = len(cdf)
    print(f"Cluster {cluster_id} | rows={size}")

    # If cluster_probability exists, show quick stats
    if "cluster_probability" in cdf.columns:
        probs = pd.to_numeric(cdf["cluster_probability"], errors="coerce")
        print(f"cluster_probability: min={probs.min():.3f} mean={probs.mean():.3f} max={probs.max():.3f}")

    # Fields to summarize
    fields = ["title_translated", "job_category", "role_k50", "role_k1500"]
    for f in fields:
        if f in cdf.columns:
            vc = cdf[f].fillna("(NA)").value_counts(dropna=False)
            print(f"\nValue counts: {f}")
            print(vc.head(10).to_string())
        else:
            print(f"\nValue counts: {f} (column missing)")

    # Show a couple of random descriptions
    if "description" in cdf.columns:
        rng = np.random.default_rng(seed)
        idxs = cdf.index.to_numpy()
        if len(idxs) > 0:
            pick = rng.choice(idxs, size=min(sample_n, len(idxs)), replace=False)
            print("\nSample descriptions:")
            for j, ridx in enumerate(pick, start=1):
                desc = str(cdf.at[ridx, "description"] or "")
                print(f"\n[{j}/{len(pick)}] index={ridx}, len={len(desc)}")
                print(desc[:1200] + ("..." if len(desc) > 1200 else ""))
        else:
            print("\nNo rows to sample descriptions from.")
    else:
        print("\nColumn 'description' not found; cannot show samples.")

    return cluster_id, cdf

# Example usage:
chosen_id, cluster_slice = inspect_random_cluster(cluster_df)
```
